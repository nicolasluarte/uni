```python
%matplotlib inline
import math
import matplotlib.pyplot as plt
import numpy as np
```

```python
def repetitions(rir, reps):
      exertion = list()
      i = 1
      for r1, r2 in zip(rir, reps):
          result = math.exp(-0.215 * (r1 + r2 - i))
          exertion.append(result)
          i += 1
      return exertion
```

```python
rir = [3, 3, 3.5, 3, 4]
reps = [8, 6, 5, 5, 4]
result = repetitions(rir, reps)
cumResult = np.cumsum(result)
```

```python
plt.plot(result)
plt.xticks(np.arange(0, len(reps), step=1))
plt.show()
```

```python
plt.plot(cumResult)
plt.xticks(np.arange(0, len(reps), step=1))
plt.show()
```
